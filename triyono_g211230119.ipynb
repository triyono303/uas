```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Purchase Price': [90.00, 32.00, 25.00, 50.00, 18.0, 225.00, 22.00, 18.00, 65.00, 800.00, 95.00, 17.00, 45.00],
    'Current Price': [144.13, 65.57, 82.57, 44.56, 17.69, 833.36, 36.07, 33.24, 56.48, 897.64, 86.26, 141.64, 14.61],
    'Gain/Loss': [14413.27, 13048.57, 12302.93, 11094.20, 8824.82, 82502.64, 7176.94, 7445.54, 10391.40, 43984.36, 8539.74, 141640.00, 7287.94]
}

df = pd.DataFrame(data)
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].hist(df['Purchase Price'], bins=10, color='blue', edgecolor='black')
axs[0].set_title('Histogram of Purchase Price')
axs[0].set_xlabel('Purchase Price')
axs[0].set_ylabel('Frequency')

axs[1].hist(df['Current Price'], bins=10, color='green', edgecolor='black')
axs[1].set_title('Histogram of Current Price')
axs[1].set_xlabel('Current Price')
axs[1].set_ylabel('Frequency')

axs[2].hist(df['Gain/Loss'], bins=10, color='orange', edgecolor='black')
axs[2].set_title('Histogram of Gain/Loss')
axs[2].set_xlabel('Gain/Loss')
axs[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
